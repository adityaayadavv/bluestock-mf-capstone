```python
import pandas as pd
from pathlib import Path

DATA = Path("../data/raw")

fund_df = pd.read_csv(DATA / "01_fund_master.csv")
nav_df = pd.read_csv(DATA / "02_nav_history.csv")

print("Fund Master:", fund_df.shape)
print("NAV History:", nav_df.shape)
```

```python
fund_df["category"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Funds by Category")
plt.show()
```

```python
fund_amfi = set(fund_df["amfi_code"])
nav_amfi = set(nav_df["amfi_code"])

missing_codes = fund_amfi - nav_amfi

print("Fund Master Codes:", len(fund_amfi))
print("NAV Codes:", len(nav_amfi))
print("Missing Codes:", len(missing_codes))
```
